```python
#4 Creating Loyalty Flags 
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
pd.options.display.max_rows = None
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# 2 Aggregate Mean of Dataframe
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

3. The results for the entire dataframe do not differ drastically, but getting the average of the full dataset is a more complete result than a subset. 

4. Loyalty Flags were already created above. 

```python
#5 Creating descriptive Statistics for Loyalty flag
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max', 'sum']})
```

```python
#6 Creating Spending Flag 
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#7 Creating Order Frequency Flag
ords_prods_merge['prior_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['prior_orders'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['prior_orders'] > 10) & (ords_prods_merge['prior_orders'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['prior_orders'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
